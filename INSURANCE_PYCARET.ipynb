```
from pycaret.datasets import get_data
data = get_data('insurance')
```

```
data['region'].unique()
```

```
from pycaret.regression import *
mod=setup(data=data,target='charges',session_id=123,normalize=True,experiment_name='insurance')
models=compare_models(n_select=4)
```

```
models
```

```
# train a voting regression specific models
gbr = create_model('gbr')
rf = create_model('rf')
blender_specific = blend_models(estimator_list = [gbr,rf])
```

```
save_model(blender_specific,"insurance_model")
```

```
plot_model(blender_specific)
```

```
predict_model(blender_specific)
```
